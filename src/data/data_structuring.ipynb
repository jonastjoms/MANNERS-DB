```python
# Sort by stamp
data["Stamp"] = pd.to_numeric(data["Stamp"])
data = data.sort_values(by='Stamp', ascending=True)
for col in data:
    data[col] = data[col].astype(float)
data['Stamp'] = data['Stamp'].astype(int)
```

```python
# Remove rows where Honeypot is wrong:
print(data.shape)
indecies = []
for i, row in data.iterrows():
    if row['Number of children'] or row['Number of animals'] > 0:
        if not row['Child or animal'] == 1:
            indecies.append(i)
    else:
        if not row['Child or animal'] == 0:
            indecies.append(i)
data = data.drop(data.index[indecies])
print(data.shape)
```

```python
# Can now remove honeypot column
data = data.drop(columns = ['Child or animal'])
display(data[data.Stamp == 986])
display(data.columns)
```

```python
# Drop nan
display(data[data.isna().any(axis=1)])
data = data.dropna()
# Save data before normalization
data.to_csv('not_normallized.csv', index=False)
# Normalize non-binary values:
binary_cols = ['Stamp', 'Using circle', 'Using arrow', 'Robot within group?', 'Robot facing group?', 'Robot facing closest human?',
            'Robot facing 2nd closest human?', 'Robot facing 3rd closest human?',
            'Closest human facing robot?', '2nd closest human facing robot?',
            '3d closest human facing robot?', 'Vacuum cleaning', 'Mopping the floor',
            'Carry warm food', 'Carry cold food', 'Carry drinks',
            'Carry small objects (plates, toys)',
            'Carry big objects (tables, chairs)',
            'Cleaning (Picking up stuff) / Starting conversation']
for col in data:
    if col not in binary_cols:
        data[col] = (data[col]-data[col].min())/(data[col].max()-data[col].min())
```

```python
# Modifying labels to get higher variance
tmp = data.copy()
tmp = tmp.reset_index()
tmp = tmp.drop(columns = ['index'])
for scene in np.arange(250,999):
    scene_df = tmp[tmp.Stamp == scene]
    if scene_df['Robot within group?'].values[0] == 1:
        len = scene_df.shape[0]
        start = scene_df.index[0] 
        stop = start+len
        tmp.iloc[start:start+len//2,-8:-1] = 5
        tmp.iloc[start+len//2:stop,-8:-1] = 1
    scene_df = tmp[tmp.Stamp == scene]
# Remember to comment this away later:
data = tmp
```

```python
# Save all data to file:
data = data.reset_index()
data = data.drop(columns = ['index'])
data.to_csv('all_data_ale.csv', index=False)

for scene in np.arange(250,999):
    if scene == 986:
        display(data[data.Stamp == scene].iloc[:,[7,-8,-7,-6,-5,-4,-3,-2,-1]])
        print(data[data.Stamp == scene].iloc[:,[7,-8,-7,-6,-5,-4,-3,-2,-1]].var())
    
# Choose 100 random scenes to be saved as test samples
data_train = data
test_scenes = np.random.randint(250, 999, size = 100)
first_it = True
for test_scene in test_scenes:
    while data_train[data_train['Stamp'] == test_scene].empty:
            test_scene +=1
    # Add scene to test dataframe:
    if first_it:
        data_test = data_train[data_train['Stamp'] == test_scene]
        first_it = False
    else:
        data_test = pd.concat([data_test, data_train[data_train['Stamp'] == test_scene]])
    # Remove selected rows from train dataframe:
    data_train = data_train[data_train.Stamp != test_scene]
```
