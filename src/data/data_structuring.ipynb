```python
# Create Header for full dataframe:
questions = ['Child or animal', 'Vacuum cleaning', 'Mopping the floor', 'Carry warm food', 'Carry cold food', 'Carry drinks', 'Carry small objects (plates, toys)', 'Carry big objects (tables, chairs)', 'Cleaning (Picking up stuff) / Starting conversation']
list = np.append(features.loc[features['Stamp'] == str(stamp)].values[0], batch.iloc[sample].values)
data = pd.DataFrame(data, columns = features.columns.values.tolist() + questions) 
data = data.replace(['Yes', 'No'], [1, 0])
display(data)
```

```python
# Sort by stamp
data["Stamp"] = pd.to_numeric(data["Stamp"])
data = data.sort_values(by='Stamp', ascending=True)
for col in data:
    data[col] = data[col].astype(float)
data['Stamp'] = data['Stamp'].astype(int)
display(data)
```

```python
# Remove rows where Honeypot is wrong:
print(data.shape)
indecies = []
for i, row in data.iterrows():
    if row['Number of children'] or row['Number of animals'] > 0:
        if not row['Child or animal'] == 1:
            indecies.append(i)
    else:
        if not row['Child or animal'] == 0:
            indecies.append(i)
data = data.drop(data.index[indecies])
print(data.shape)
```

```python
# Can now remove honeypot column
data = data.drop(columns = ['Child or animal'])
display(data)
display(data.columns)
```

```python
# Normalize non-binary values:
binary_cols = ['Stamp', 'Using circle', 'Using arrow', 'Robot within group?', 'Robot facing group?', 'Robot facing closest human?',
            'Robot facing 2nd closest human?', 'Robot facing 3rd closest human?',
            'Closest human facing robot?', '2nd closest human facing robot?',
            '3d closest human facing robot?', 'Vacuum cleaning', 'Mopping the floor',
            'Carry warm food', 'Carry cold food', 'Carry drinks',
            'Carry small objects (plates, toys)',
            'Carry big objects (tables, chairs)',
            'Cleaning (Picking up stuff) / Starting conversation']
for col in data:
    if col not in binary_cols:
        data[col] = (data[col]-data[col].min())/(data[col].max()-data[col].min())
display(data)
```

```python
# Drop nan
display(data[data.isna().any(axis=1)])
data = data.dropna()
```

```python
# Choose 100 random scenes to be saved as test samples
print(data.shape)
test_scenes = np.random.randint(999, size = 100)
first_it = True
for test_scene in test_scenes:
    # Add scene to test dataframe:
    if first_it:
        data_test = data[data['Stamp'] == test_scene]
        first_it = False
    else:
        data_test = pd.concat([data_test, data[data['Stamp'] == test_scene]])
    # Remove selected rows from train dataframe:
    data = data[data.Stamp != test_scene]
print(data.shape)
```
