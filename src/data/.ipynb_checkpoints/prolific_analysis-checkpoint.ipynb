```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '/Users/jonastjomsland/UCB/dissertation/data/prolific1.csv'
fig_path = '/Users/jonastjomsland/UCB/dissertation/src/data_cleaning/answers/'
```

```python
df = pd.read_csv(data_path)
pd.set_option('max_columns', None)
```

```python
df.drop(df.tail(2).index,inplace=True)
#display(df)
```

```python
# Dictionary for all answers:
dict = {}
for (columnName, columnData) in df.iteritems():
    if columnName[0] == 'Q':
        scene = columnName[0:-4]
        question = columnName[-1]
        if scene not in dict:
            dict[scene] = {}
            dict[scene][question] = {}
            dict[scene][question]['data'] = columnData.values[2:]
            if columnData.values[0][83:89] == 'circle':
                dict[scene][question]['title'] = columnData.values[0][131:]
            else: 
                dict[scene][question]['title'] = columnData.values[0][141:]
        if question not in dict[scene]:
            dict[scene][question] = {}
            dict[scene][question]['data'] = columnData.values[2:]
            if columnData.values[0][83:89] == 'circle':
                dict[scene][question]['title'] = columnData.values[0][131:]
            else: 
                dict[scene][question]['title'] = columnData.values[0][141:]
```

```python
def bar_plot(answer, scene, question, title):
    d = {}
    for element in answer:
        if element not in d:
            d[element] = 1
        else:
            d[element] += 1
    x = range(5)
    labels = ['Very inappropriate', 'Inappropriate', 'Neutral', 'Appropriate', 'Very appropriate']
    count = np.zeros((5))
    for i in range(5):
        try:
            count[i] = d[labels[i]]
        except KeyError:
            count[i] = 0
    plt.figure()
    plt.bar(x, count, align='center')
    plt.xticks(x, x)
    plt.title(title, fontsize = 16)
    plt.savefig(fig_path+scene+'/'+question+'.png')
    return True
        
```

```python
for scene in dict:
    for question in dict[scene]:
        answer = dict[scene][question]['data']
        title = dict[scene][question]['title']
        bar_plot(answer, scene, question, title)
```
